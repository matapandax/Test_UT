```python
 #Reveresed String in Python
 def reversed_string(text):
     result = ""
     index = len(text) - 1
     while index >= 0:
         result += text[index]
         index -= 1
     return result

#Driver Code
 reversed_string("Warta Niaga")
```

```python
# Recursive function to check if a
# string is palindrome

def isPalindrome(s):

    # to change it the string is similar case
    s = s.lower()
    # length of s
    l = len(s)

    # if length is less than 2
    if l < 2:
        return True

    # If s[0] and s[l-1] are equal
    elif s[0] == s[l - 1]:

        # Call is pallindrome form substring(1,l-1)
        return isPalindrome(s[1: l - 1])

    else:
        return False


# Driver Code
s = input("enter string :  ")
ans = isPalindrome(s)

if ans:
    print("True")

else:
    print("False")
```

```python
# Python program to check if year is a leap year or not

year = 2024

# To get year (integer input) from the user
# year = int(input("Enter a year: "))

if (year % 4) == 0:
   if (year % 100) == 0:
       if (year % 400) == 0:
           print("{0} Ini Tahun Kabisat".format(year))
       else:
           print("{0} Ini Bukan Tahun Kabisat".format(year))
   else:
       print("{0} Ini Tahun Kabisat".format(year))
else:
   print("{0} Ini Bukan Tahun Kabisat".format(year))
```

```python
start = int(input("Enter start year: "))
end = int(input("Enter end year: "))

while start < end:
    if start % 4 == 0:
        if start % 100 == 0:
            if start % 400 == 0:
                print("{} Ini Tahun Kabisat".format(start))
            else:
                print("{} Ini Bukan Tahun Kabisat".format(start))
        else:
            print("{} Ini Tahun Kabisat".format(start))
    else:
        print("{} Ini Bukan Tahun Kabisat".format(start))
    start += 1
```
